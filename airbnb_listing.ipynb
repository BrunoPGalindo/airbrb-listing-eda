```python
# seeing how far the outlier is

sns.boxplot(data=data, x='price')
plt.show()
```

```python
# there are outliers! we need a better graph
```

```python
# new dataframe with lower values to be represented
```

```python
df_price = data[data['price'] < 1500]

# price distribution
sns.histplot(data=df_price, x='price', bins=120)
plt.title('Price distribution')
plt.ylabel('Frequency')
plt.xlabel('Price')
plt.show()
```

```python
data.dtypes
```

```python
sns.histplot(data=df_price, x='availability_365')
plt.title('Availability throughout the year')
plt.ylabel('Frequency')
plt.xlabel('Quantity of days')
plt.show()
```

```python
data.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df_price
```

```python
data.columns
```

```python
sns.barplot(data=df_price, x='neighbourhood_group', y='price', hue='room_type')
plt.ylabel('Price')
plt.xlabel('Location')
plt.show()
```

```python
sns.scatterplot(data=df_price, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.title('Reviews by location')
plt.ylabel('Price')
plt.xlabel('Number of reviews')
plt.show()
```

```python
sns.pairplot(data=df_price, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
plt.show()
```

```python
# geografic distribution
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df_price, x='longitude', y='latitude', hue='room_type')
plt.title('Geografic Distribution')
plt.ylabel('Latitude')
plt.xlabel('Longitude')
plt.show()
```

```python
# heatmap - correlation
```

```python
corr = df_price[['latitude', 'longitude', 'price', 'minimum_nights'
                 , 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
```

```python
sns.heatmap(data=corr, annot=True)
plt.show()
```
